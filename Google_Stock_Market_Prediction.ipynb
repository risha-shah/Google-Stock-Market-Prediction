```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import  MinMaxScaler
```

```python
from google.colab import files
data = files.upload()
```

```python
data = pd.read_csv ('GOOG.csv', date_parser= True)
data.tail()
```

```python
#split the data 
data_training = data[data['Date'] < '2021-01-01'].copy() #exlcuding 2018 stock prices, used for testng data
data_training
```

```python
#split the data  to testing data
data_testing = data[data['Date'] >= '2018-01-01'].copy() #exlcuding 2018 stock prices, used for testng data
data_testing
```

```python
training_data = data_training.drop(['Date', 'Adj Close'], axis = 1)
training_data.head(5)
```

```python
scaler = MinMaxScaler()
training_data = scaler.fit_transform(training_data)
training_data
```

```python
X_train = []
Y_train = []
```

```python
training_data.shape[0]
```

```python
for i in range (60, training_data.shape[0]):
    X_train.append(training_data[i-60: i]) 
    Y_train.append(training_data[i,0])
```

```python
X_train, Y_train = np.array(X_train), np.array(Y_train)
```

```python
X_train.shape
```

```python
Y_train.shape 
```

```python
#BUILDING THE LSTM
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout 

```

```python

regressior = Sequential()
regressior.add(LSTM(units=50, activation= "relu", return_sequences= True, input_shape = (X_train.shape[1] , 5))) #THE 2 INPUT PARAMETERS, how many units/layers you want in LSTM network
regressior.add(Dropout(0.2)) 

regressior.add(LSTM(units=60, activation= "relu", return_sequences= True, input_shape = (X_train.shape[1] , 5))) #THE 2 INPUT PARAMETERS, how many units/layers you want in LSTM network
regressior.add(Dropout(0.3)) 

regressior.add(LSTM(units=80, activation= "relu", return_sequences= True, input_shape = (X_train.shape[1] , 5))) 
regressior.add(Dropout(0.4)) 

regressior.add(LSTM(units=120, activation= "relu"))
regressior.add(Dropout(0.5)) 
regressior.add(Dense(units = 1))
```

```python
X_train.shape[1] , 5 
```

```python
regressior.summary()
```

```python
regressior.compile(optimizer= "adam", loss = "mean_squared_error") 
```

```python
regressior.fit(X_train, Y_train, epochs=10, batch_size = 32)
```

```python
df= past_60_days.append (data_testing, ignore_index= True)
df 
```

```python
df = df.drop (['Date', 'Adj Close'], axis =1)
df.head()
```

```python
inputs = scaler.transform(df)
inputs 
```

```python
X_test = []
y_test = []

for i in range (60, inputs.shape[0]):
  X_test.append (inputs[i-60:i])
  y_test.append (inputs[i,0]) #represents the opening stock price
```

```python
X_test, y_test = np.array (X_test), np.array(y_test)
X_test.shape, y_test.shape
```

```python
y_pred= regressior.predict (X_test)
```
